```python
import random
```

```python
rows, cols = (4, 4)
arr=[[0 for i in range(cols)] for i in range(rows)]

for i in arr:
  print(i)
```

```python
for row in range(4):
    for col in range(4):
        arr[row][col]=random.random() 
for row in range(4):
    print(arr[row])        
```

```python
Tarr=[[0 for i in range(cols)] for i in range(rows)]
for row in range(4):
    for col in range(4):
        Tarr[row][col]=arr[col][row]
        
for row in range(4):
    print(Tarr[row])
        
```

```python
mularr=[[0 for i in range(cols)] for i in range(rows)]
for i in range(4):
    for j in range(4):
        sum=0
        for k in range(4):
            sum+=arr[i][k]*Tarr[k][j]
        mularr[i][j]=sum 
         
for row in range(4):
    print(mularr[row])
```
