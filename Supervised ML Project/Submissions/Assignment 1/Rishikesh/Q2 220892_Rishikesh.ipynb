```python
#Drawing curve of sinx using numpy and matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,50)
y = np.sin(x)
plt.plot(x,y)
```

```python
#Drawing the percentage share of subjects in my CPI
TA = 5
ESC = 7
CHM = 10
PHY = 8
LIF = 9
PE = 10

import matplotlib.pyplot as plt

# Data
subjects = ['TA', 'ESC', 'CHM', 'PHY', 'LIF', 'PE']
percentages = [TA, ESC, CHM, PHY, LIF, PE]

# Plotting
plt.pie(percentages, labels=subjects, autopct='%1.1f%%', startangle=90)

# Title
plt.title('Percentage Share of Subjects in CPI')

# Display the graph
plt.show()
```

```python
# Drawing the comaparison of marks in my JEE Mains and Advaned
import matplotlib.pyplot as plt

# representing Subjects using abbreviation PM(physics mains)
PM = 90
PA = 47
CM = 86
CA = 72
MM = 100
MA = 45
subjects = ['PM', 'PA', 'CM', 'CA', 'MM', 'MA']
marks = [ PM, PA, CM, CA, MM, MA]

# Plotting
plt.bar(subjects, marks)

# Labels and title
plt.xlabel('Subjects')
plt.ylabel('Percentage')
plt.title('comaparison of marks in my JEE Mains and Advaned')

# Display the graph
plt.show()
```
